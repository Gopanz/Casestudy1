# 3.Bar Graph (male vs female)

```python
dataset['Buyer Gender'].value_counts()
```

```python
plt.figure(figsize=(7,7))
buyer = dataset['Buyer Gender'].value_counts().to_frame().reset_index()
x = buyer['index']
y = buyer['Buyer Gender']
plt.bar(buyer['index'],buyer['Buyer Gender'],color = ['blue','green'])
plt.title('Gender Count',fontsize=15)
plt.xlabel('Gender')
plt.ylabel('Count') 
```
